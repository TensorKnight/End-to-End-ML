### 2.2 Dataset Description

This dataset contains daily climate and geographic features for multiple areas. It can be used for tasks such as climate modeling, habitability prediction, environmental classification, or geographical impact analysis. Below is the detailed description of each column:

* **Area Name**: Unique identifier for each geographical area.
* **Daily Min Temp**: Minimum daily temperature (in °C).
* **Daily Max Temp**: Maximum daily temperature (in °C).
* **Daily Avg Temp**: Average daily temperature (in °C).
* **Total Precipitation**: Total daily precipitation (in mm).
* **Relative Humidity**: Average relative humidity (in %).
* **Water Bodies**: Binary flag indicating presence (1) or absence (0) of nearby water bodies.
* **Population Density**: Number of people per square kilometer.
* **Day Length**: Duration of daylight in hours.
* **Urban / Rural Area**: Categorical binary flag where 1 = Urban, 0 = Rural.
* **Forested Area**: Binary indicator of presence of forested land (1 = Yes, 0 = No).
* **Crop Area**: Binary indicator for presence of agricultural land.
* **Graze Land Area**: Binary indicator for grazing land presence.
* **Habitable**: Target variable — indicates whether the area is habitable (1 = Yes, 0 = No).

### 3.7 Feature Engineering

#### 3.7.0 Natural resource score

```python
df['Natural Resource Score'] = df[['Forested Area', 'Water Bodies', 'Crop Area', 'Graze Land Area']].sum(axis=1)
```

#### 3.7.1 Humidity Temprature Score 

```python
df['Humidity_Temp'] = df['Relative Humidity'] * df['Daily Avg Temp']
```

### 3.8 Visualization

#### 3.8.0 Numerical Feature distribution

```python
num_cols = df.select_dtypes(include='number').columns.tolist()
df[num_cols].hist(figsize=(15, 10), bins=30, edgecolor='black')
plt.suptitle('Distributions of Numerical Features', fontsize=16)
plt.tight_layout()
plt.show()
```

Insight :

1. **Daily Min Temp** – Normally distributed, useful as-is for modeling.
2. **Daily Max Temp** – Bell-shaped curve, strong candidate without transformation.
3. **Daily Avg Temp** – Symmetric distribution, likely a key predictor.
4. **Total Precipitation** – Uniformly distributed; binning may improve interpretability.
5. **Relative Humidity** – Evenly spread, may interact meaningfully with temperature.
6. **Water Bodies** – Highly imbalanced binary, needs resampling or rare-event handling.
7. **Population Density** – Wide range; log transformation could help reduce skew impact.
8. **Day Length** – Fairly uniform; may influence seasonal habitability.
9. **Urban / Rural Area** – Binary and imbalanced; check for predictive power.
10. **Forested Area** – Rare binary feature; combine or drop if not informative.
11. **Crop Area** – Skewed binary; may be more meaningful when grouped with others.
12. **Graze Land Area** – Imbalanced; could be merged into a broader land use feature.
13. **Habitable** – Binary target with imbalance; modeling should address class imbalance.


#### 3.8.1 Boxplot for Outlier Detection

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

n_cols = 3 
n_rows = (len(num_cols) + n_cols - 1) // n_cols  

plt.figure(figsize=(n_cols * 5, n_rows * 4))

for i, col in enumerate(num_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

Insight :

No outlier as such 

#### 3.8.2 Correlation features with target variable

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64'])

corr_target = numerical_cols.corr()['Habitable'].sort_values(ascending=False)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_target.to_frame(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation of Features with Habitable')
plt.show()
```

Insight : 
Water bodies and Natural Resource Score has the highest correlation with the target feature
